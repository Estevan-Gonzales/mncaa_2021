```python
training_15 = pd.read_csv('MRegularSeasonCompactResults.csv')
training_15 = training_15[training_15['Season'] < 2016]
#training_15 = training_15[training_15['Season'] > 2010]
training_15.head()
```

```python
matchups = {}

wteam = training_15.WTeamID.tolist()
lteam = training_15.LTeamID.tolist()

for i in range(len(wteam)):
    if wteam[i] in matchups:
        if lteam[i] in matchups[wteam[i]]:
            matchups[wteam[i]][lteam[i]] += 1
        else:
            matchups[wteam[i]][lteam[i]] = 0
    else:
        matchups[wteam[i]] = {lteam[i]: 1}
        
    
```

```python
win_perc = {}

for winner in matchups:
    for loser in matchups[winner]:
        try:
            wins = matchups[winner][loser]
        except:
            wins = 0
        try:
            losses = matchups[loser][winner]
        except:
            losses = 0
        total = wins + losses
        
        try:
            win_perc[str(winner) + '_' + str(loser)] = wins/total
        except:
            win_perc[str(winner) + '_' + str(loser)] = .5
        try:
            win_perc[str(loser) + '_' + str(winner)] = losses/total
        except:
            win_perc[str(loser) + '_' + str(winner)] = .5
    
```

```python
preds = []

for game in sample.ID.tolist():
    game_split = game.split('_')
    team1 = game_split[1]
    team2 = game_split[2]
    
    try:
        preds.append(win_perc[str(team1) + '_' + str(team2)])
    except:
        try:
            preds.append(1 - win_perc[str(team2) + '_' + str(team1)])
        except:
            preds.append(.5)
```

```python
import numpy as np
sample['Pred'] = np.array(preds)
```
